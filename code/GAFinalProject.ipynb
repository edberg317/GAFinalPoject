## Analysis:

At this stage, the DataFrame should be ready for the next steps which are visualization and analysis.

1. Get an overview of the descriptive statistics of all the columns.

2. Visualize the distribution of cancer patients' age in histogram. 
[Go to Figure 1 interpretation.](#figure1_i)
<a id="figure1"></a>

```python
#by knowing that the min and max age of cancer patients are 14 and 73 respectively,
#the histogram is designed to have bins=8 and a range between 0-80

plt.hist(cancer_patients['Age'], bins=8, range=(0,80))
plt.ylabel('Number of Cancer Patients')
plt.xlabel('Cancer Patients\' Age')
plt.title('Figure 1: Distribution of Cancer Patients\' Age');
```

3. Visualize the distribution of cancer patients of various age groups in pie chart.

```python
cancer_patients['AgeGroup'] = cancer_patients['Age'].apply(lambda x: int((x/10))+1)
# categorize the age of cancer patients into various groups
# 1-10 = 1
# 11-20 = 2
# 21-30 = 3
# 31-40 = 4
# 41-50 = 5
# 51-60 = 6
# 61-70 = 7
# 71-80 = 8
```

```python
cancer_patients[['Age', 'AgeGroup']].sort_index()
#display the Age and AgeGroup columns to make sure that the cancer patients' age is categorized properly
```

```python
cancer_patients.groupby(['AgeGroup']).count()
#display the number of cancer patients in various age group
```

[Go to Figure 2 interpretation.](#figure2_i)
<a id="figure2"></a>

```python
age_group = ('11-20 y/o', '21-30 y/o', '31-40 y/o', '41-50 y/o', '51-60 y/o', '61-70 y/o', '71-80 y/o')
#create a tuple of pie chart labels of different age group
             
plt.pie(cancer_patients.groupby(['AgeGroup']).count()['Age'], labels = age_group, autopct='%.1f%%',\
        shadow=True,explode=(0.3,0.1,0.1,0.1,0.3,0.3,0.3))
plt.title('Figure 2: Distribution of Cancer Patients\' Age in Various Age Groups');
```

4. Visualize the number of cancer patients of different severity level in bar graph. [Go to Figure 3 interpretation](#figure3_i)
<a id="figure3"></a>

```python
cancer_patients['Level'].value_counts().sort_values(ascending=True).plot(kind='bar')
plt.ylabel('Number of Cancer Patients')
plt.xlabel('Severity Level')
plt.title('Figure 3: Number of Cancer Patients of Different Severity Level');
```

5. Visualize the number of cancer patients of different severity level of various age groups in stacked bar graph. 

```python
#define a function that can retrieve data array of different levels 
def data_array_of_different_level(level, column_name, *column_data_range):
    '''
    parameter level could be 'Low', 'Medium', 'High'
    parameter column_name could be any elements in df.columns
    parameter column_data_range is tuple(range(df[column].min(),df[column].max()+1))
    '''
    result = []
    for i in column_data_range:
        result.append(cancer_patients[(cancer_patients['Level']==level) & (cancer_patients[column_name]==i)].shape[0])
    return np.array(result)
```

```python
cancer_patients.groupby(by=['Level', 'AgeGroup'])[['Age']].count()
#number of cancer patients are group by level and age group
```

```python
age_group_range = (2,3,4,5,6,7,8)
age_group_low = data_array_of_different_level('Low','AgeGroup',2,3,4,5,6,7,8)
age_group_medium = data_array_of_different_level('Medium','AgeGroup',2,3,4,5,6,7,8)
age_group_high = data_array_of_different_level('High','AgeGroup',2,3,4,5,6,7,8)
```

6. Verify the relationship between cancer patient's age and severity level. [Go to Figure 5 interpretation](#figure5_i)
<a id="figure5"></a>

```python
x = cancer_patients[cancer_patients['Level']=='Low'][['Age']]
y = cancer_patients[cancer_patients['Level']=='Medium'][['Age']]
z = cancer_patients[cancer_patients['Level']=='High'][['Age']]
plt.boxplot(x, positions =[1], widths = 0.2)
plt.boxplot(y, positions =[2], widths = 0.2)
plt.boxplot(z, positions =[3], widths = 0.2)
plt.xticks([1, 2, 3], ['Low', 'Medium', 'High'])
plt.ylabel('Cancer Patient\'s Age')
plt.xlabel('Severity Level')
plt.title('Figure 5: Relationship Between Severity Level and Cancer Patient\'s Age');
```
